```python
# Replace this variable with the path to your folder containing video files
folder_path = "PATH DIRECTIONS HERE"

# List all files in the folder
files = os.listdir(folder_path)

# Filter only video files (you can add more video extensions if needed)
video_files = [file for file in files if file.lower().endswith(('.mp4', '.avi', '.mkv', '.mov'))]

# Loop through each video file and get its duration
for video_file in video_files:
    video_path = os.path.join(folder_path, video_file)
    video = VideoFileClip(video_path)
    video_duration = video.duration
    video.close()
    print(f"Video: {video_file} - Duration: {video_duration:.2f} seconds")
```
